```python
#Sentiment analysis using transformers(detailed processing)
#step by step implementation

#Step1:Install necessary libraries
#Ensure you have the transformers and related librarries installed
#Run pip install
```

```python
from transformers import AutoTokenizer,AutoModelForSequenceClassification
import torch
import numpy as np
```

AutoTokenizer:loads a tokenizer that converts text into tokenizer suitable for the model

AutoModelForSequenceClassification:

```python
#Step 2:load the pre trained midel and tokenizer
model_name="distilbert-base-uncased-finetuned-sst-2-english"
tokenizer=AutoTokenizer.from_pretrained(model_name)
model=AutoModelForSequenceClassification.from_pretrained(model_name)
```

model_name:specifies a model pretrained on sentiment analysis(SST2 dataset.

AutoTokenizer.from_pretrained:loads the tokenizer for the specified model

AutoModelForSequenceClassification.from_pretrained:loads the mpodel fine tuned for binary sentiment classification



```python
#Step3 tokenization and input embedding
texts=[
    "I love this product! It's amazing",
    "This was the worst experinece ever.Terrible service",
    "The movie was okay,not great but not bad either ",
    "Absouletly fantasctic! Couldn't be happier",
    "Meh! it was average at best"

]
```

```python
tokens=tokenizer(texts,padding=True,truncation=True,return_tensors="pt")
input_ids=tokens['input_ids']
attention_mask=tokens['attention_mask']

```

tokenizer:converts text into tokens,adding special tokens like CLS and SEP

padding:ensures all  the sequences are of the same lemgth

truncation:trims the sentence that are too long

return_tensors="pt":returns tensord compatible with pytorch

input_ids:encoded token ids for each sentence

binary mask 1 for good 0 for bad

```python

#Step4-7:Transformers layers abstract
#4:positional encode
#5: self attention mechnism
#6:add and norm layers
#7: feed forward networks
#these steps describe the internal processes within the transformers

#positional encoding ensures order unformation is preserved
#self attemtion mechanism calculates how each token relates to ecah tother
#add and norms appies residual connections and normalizatin to stablize training
#feed forward networks refines token representation with dense layers
```

```python
#Step 8:classification output
with torch.no_grad():
  logits=model(input_ids,attention_mask=attention_mask).logits
```

```python
#Step 9:softmax for probability distribution
probs=torch.nn.functional.softmax(logits,dim=-1)
sentiment_labels=["NEGATIVE","POSITIVE"]
predictions=[sentiment_labels[torch.argmax(prob).item()] for prob in probs]
```

softmax converts logits to probablities

sentiment_lables defines class lables

torch.argmax: finds the index of the highest probability

predoctions maps the highest probabiltiy to a sentiment label

```python
#Step 10: display results
for text,prob,prediction in zip(texts,probs,predictions):
  confidence=torch.max(prob).item()
  print(f"Text:{text}")
  print(f"Sentiment:{prediction},Confidence:{confidence:.2f}")
  print("-")
```

zip combines texts,probabilities and predictions for iteration

torch.max finds the confidemce score for the predicyion

print displays eact text its sentiment and the confidence score
